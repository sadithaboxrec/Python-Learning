```python
# To count to 10

for x in range (1,11):
    print(x)
```

```python
# reversed
for x in reversed(range (1,11)):
    print(x)
print("Happy New Year!")
```

```python
# To count to 10 with 2 steps between each

for x in range (1,11,2):
    print(x)
```

```python
credit_card="1234-5678"

for x in credit_card:
    print(x)
```

```python
for x in range (1,11):
    if x  == 6:
        continue
    else:
        print(x)
```

```python
for x in range (1,11):
    if x  == 6:
        break
    else:
        print(x)
```
