### Libraries

##### Libraries are generally files of code that written we can use in our own program or we wriiten.

#### Modules in python just library that typically has one or more functions or built into it.

#### import allows you to import functions from some modules in python

```python
import random

coin= random.choice(["heads","tails"])
print(coin)
```

#### from keyword can be used when importing a function from module but it allows to be little more specific

```python
from random import choice
coin = choice(["heads","tails"])
print(coin)
```

```python
from random import randint
 # randint(a,b)  numbers from a to b
number=randint(1,10)
print(number)
```

```python
from random import shuffle

cards=["jack","queen","king","ace"]
shuffle(cards)

for card in cards:
    print(card)
```
