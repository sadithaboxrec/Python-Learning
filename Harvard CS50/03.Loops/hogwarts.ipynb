# Lists

```python
students=["Hermione","Harry","Ron"]
for student in students:
    print(student)

# python initialize student variable, no need to do it separately
```

```python
students=["Hermione","Harry","Ron"]
for student in students:
    print(students[0])

# students[0] is always "Hermione".
#The loop runs 3 times, so it prints Hermione three times.
```

```python
students=["Hermione","Harry","Ron"]
for student in students:
    print(student[0])
```

```python
students=["Hermione","Harry","Ron"]

# len function help to dynamically figure out how many elements inside it

for i in range(len(students)):
    print(students[i])
```

```python
students=["Hermione","Harry","Ron"]

for i in range(len(students)):
    print(i,students[i])
```

```python
students=["Hermione","Harry","Ron"]

for i in range(len(students)):
    print(i+1,students[i])
```

# dict

```python
students={"Hermione":"Gryffindor", "Harry":"Gryffindor","Ron":"Gryffindor","Draco":"Slytherin"}

print(students["Hermione"])
```

```python
students={"Hermione":"Gryffindor", "Harry":"Gryffindor","Ron":"Gryffindor","Draco":"Slytherin"}

for student in students:
    print(student)
```

```python
students={"Hermione":"Gryffindor", "Harry":"Gryffindor","Ron":"Gryffindor","Draco":"Slytherin"}

for student in students:
    print(students)
```

```python
students={"Hermione":"Gryffindor", "Harry":"Gryffindor","Ron":"Gryffindor","Draco":"Slytherin"}

for student in students:
    print(student,"-", students[student])
```

```python
students={"Hermione":"Gryffindor", "Harry":"Gryffindor","Ron":"Gryffindor","Draco":"Slytherin"}

for student in students:
    print(student, students[student] ,sep="-")
```

```python
students = [
    {"name":"Hermione","house":"Gryffindor","patronus":"Otter"},
    {"name":"Harry","house":"Gryffindor","patronus":"Stag"},
    {"name":"Ron","house":"Gryffindor","patronus":"Jack Russel Terrier"},
    {"name":"Draco","house":"Slytherin","patronus":None},
]
for student in students:
    print(student["name"],student["house"],student["patronus"],sep="-")
```
