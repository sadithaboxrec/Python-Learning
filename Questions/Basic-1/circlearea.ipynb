## Write a Python program that calculates the area of a circle based on the radius entered by the user.

####                     FlowChart

##### Get the input , covert it to float and assign it to variable
                  ||
##### check if it is larger than 0 or ask the question again
                  ||
##### calculate area
                  ||
##### print



```python
while True:
    radius = float(input("Enter the radius of circle: "))

    if radius < 0:
        continue

    area = 3.1416 * radius **2
    print(f"The area of the circle with radius {radius} is: {area}")
    print(f"The area is {area:.2f}")

    break
```

## Write a Python program that calculates the circumference of a circle based on a given radius.

```python
while True:
    radius = float(input("Enter the radius of circle: "))
    if radius < 0:
        continue

    circumference = 2 * pi * radius
    print(f"The circumference of the circle with radius {radius} is: {circumference}")
    print(f"Circumference is: {circumference:0.2f}")
    break
```

## Write a Python script that takes the diameter of a circle as input and calculates its area.

```python
while True:
    diameter = float(input("Enter the diameter of circle: "))
    if diameter < 0:
        continue

    area = (pi * pow(diameter, 2)) /4
    print(f"The area of the circle with diameter {diameter} is: {area}")
    print(f"The area is {area:.2f}")

    break
```

## Write a Python program to calculate the area of a sector given the radius and angle in degrees.

```python
# radius = 3
# angle = 90
# area 7.0686

while True:
    radius = float(input("Enter the radius of sector: "))

    if radius < 0:
        continue

    angle = float(input("Enter the angle of sector: "))

    if angle < 0 or angle > 360:
        continue

    area = (pi * pow(radius, 2) * angle) /360

    print(f"The area of the sector with  {angle} angle and {radius} is: {area}")
    print(f"The area is {area:.2f}")
    break
```
