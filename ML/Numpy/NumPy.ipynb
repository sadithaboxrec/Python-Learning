#### List vs Numpy- Time Taken

```python
from time import process_time
```

##### Time Taken by a list

```python
python_list=[i for i in range(10000000)]

start_time=process_time()
python_list=[i+5 for i in python_list]
end_time=process_time()

time = end_time-start_time
print(time)
```

##### Time Taken by a Numpy Array

```python
np_array=np.array([i for i in range(10000000)])

start_time=process_time()
np_array +=5
end_time=process_time()
np_time = end_time-start_time
print(np_time)
```

### Initial placeholders in NumPy arrays

```python
# create a numpy arrays of 0s

x=np.zeros((4,5))
print(x)
```

```python
# create a numpy arrays of ones

y=np.ones((3,3))
print(y)
```

```python
# array of a particular values
z=np.full((5,4),5)    # matric shape , value
print(z)
```

```python
# create an identity matrix

a=np.eye(4)
print(a)
```

###  create a numpy array with random values

```python

# value with between 0 and 1
b=np.random.random((3,4))      # shape of the matric
print(b)
```

```python
# random integer array within a specific range

c=np.random.randint(10,100,(3,4))     #(start value, end value, matric size)
print(c)
```

```python
# array of evenly spaced values  ---> Specifying the number of values required

d=np.linspace(10,30,6)   #(start range,end range,number of values we need)
print(d)


"""
✅ 1. np.linspace(start, stop, num)
Creates num evenly spaced values between start and stop
Includes the stop value
"""
```

```python
# array of evenly spaced values  ---> specifying the step
e=np.arange(10,30,5)   #(start range,end range,)
print(e)

"""
✅ 2. np.arange(start, stop, step)

Creates values starting at start
Increments by step
Does NOT include the stop value
Similar to Python’s range()

"""
```

### Array Manipulation

```python
array=np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose   - reversing the array shape

trans =np.transpose(array)
print(trans)

print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping array

a=np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b=a.reshape((3,2))
print(b)
print(b.shape)
```
