```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Seaborn has some in built databases


```python
# total bill vs tip dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# setting a theme for the plots

#sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips, y="tip", x="total_bill", col="time" ,hue="smoker",style="smoker" ,size="size")
```

```python
# iris dataset

iris = sns.load_dataset("iris")
```

```python
iris.head()
```

#### scatter plot

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
```

```python
sns.scatterplot(x="sepal_length", y="petal_width", hue="species", data=iris)
```
