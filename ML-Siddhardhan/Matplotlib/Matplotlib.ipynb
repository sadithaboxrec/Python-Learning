```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# x = np.linspace(-np.pi, np.pi, 100)

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

## Plotting the Data

```python
#sine wave
plt.plot(x, y)
plt.show()
```

```python
#cosaine wave
plt.plot(x, z)
plt.show()
```

```python
# adding title x and y axis
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

## Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])   #ax=axes
languages=['English','French','Russian','Japanese','Chinese']
people=[100,50,150,50,30]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people amount')
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Russian','Japanese','Chinese']
people=[100,50,150,50,30]
ax.pie(people,labels=languages,autopct='%1.1f%%')   #autopct tells how many floating and decimal points we need
plt.show()
```

## Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)

fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])

ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

## 3D Scatter Plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z =20 * np.random.random(100)
x =np.sin(z)
y =np.cos(z)
ax.scatter(x,y,c=z,cmap='Reds')
plt.show()
```
